```python
# Use a file that has been uploaded, read into pandas dataframe.
import io
import pandas as pd

df3 = pd.read_csv(io.BytesIO(uploaded['austinHousingDataRaw.csv']))
```

```python
# For SQL join with price appreciation data, modify raw file for loading to sql
pd.set_option('display.max_columns', None)

# Strip day and hyphen from latest_saledate to mach appreciation data table
df3['month_year'] = df3['latest_saledate'].str[:-3]

# Add date column to sort for earliest and latest sale
df3['latest_saledate_dt'] = pd.to_datetime(df3['latest_saledate'], format='%Y-%m-%d')
df3.sort_values(by='latest_saledate_dt', inplace=True)

```
